```
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```
models = ["DS-RF", "DS_LR", "DS-SVM", "SMOTE-RF", "SMOTE-LR", "SMOTE-SVM", "BRF", "WRF", "CS-SVM"]
costs = [11540, 21990, 24200, 18700, 17860, 13000, 11330, 27730, 13920]
f1_scores = [0.462, 0.54, 0.508, 0.725, 0.686, 0.514, 0.468, 0.825, 0.46]
precisions = [0.302, 0.384, 0.353, 0.63, 0.547, 0.356, 0.307, 0.889, 0.302]
recalls = [0.984, 0.912, 0.904, 0.854, 0.92, 0.925, 0.984, 0.77, 0.971]
```

```
x = models
y = costs

fig, ax = plt.subplots()    
width = 0.6 # the width of the bars 
ind = np.arange(len(y))  # the x locations for the groups
ax.barh(ind, y, width, color="purple")

ax.set_yticks(ind+width/2)
ax.set_yticklabels(x, minor=False)
ax.set_xlim(0, 30600)
plt.title('Model comparison by cost')
plt.xlabel('costs') 

for i, v in enumerate(y):
    ax.text(v + 3, i, str(v))

plt.show()
```

```
labels = models

x = np.arange(len(labels))  # the label locations
width = 0.3  # the width of the bars

fig, ax = plt.subplots()

fig.set_figheight(5)
fig.set_figwidth(15)

rects1 = ax.bar(x - width, f1_scores, width, label='f1-score', color = "indianred", edgecolor = "white")
rects2 = ax.bar(x , precisions, width, label='precision', color="seagreen", edgecolor = "white")
rects3 = ax.bar(x + width,recalls, width, label='recall', color="royalblue", edgecolor = "white")

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_title("Model comparison")
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()


def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')


autolabel(rects1)
autolabel(rects2)
autolabel(rects3)

fig.tight_layout()

plt.show()
```
